Show stochastic turbulent wind simulation results

```python
import pandas as pd
import numpy as np
from helpers import read_out
```

```python
dfs = []
for i in range(10):
    df = read_out(f"../runs/baseline/seed{i:02d}/5MW_Land_DLL_WTurb.out")
    df["GenSpeed"] *= np.pi / 30
    dfs.append(df)
```

Just one simulation:

```python
dfs[0][["Wind1VelX", "GenTq", "GenSpeed", "BlPitchC1"]].loc[:].plot(subplots=True, figsize=(8, 8));
```

All the realisations of the stochastic turbulent wind field:

```python
ax = dfs[0][["Wind1VelX", "GenTq", "GenSpeed", "BlPitchC1", "TwrBsMyt"]] \
    .loc[:].plot(color="k", alpha=0.2, subplots=True, figsize=(8, 8))
for df in dfs:
    df["Wind1VelX"].plot(color="k", alpha=0.2, ax=ax[0])
    df["GenTq"].plot(color="k", alpha=0.2, ax=ax[1])
    df["GenSpeed"].plot(color="k", alpha=0.2, ax=ax[2])
    df["BlPitchC1"].plot(color="k", alpha=0.2, ax=ax[3])
    df["TwrBsMyt"].plot(color="k", alpha=0.2, ax=ax[4])
```
